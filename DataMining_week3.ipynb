```python
df = pd.read_csv('train.csv')
df.describe() # csv dosya içeriği
```

```python
df_cleaned = df.drop(['Survived', 'PassengerId', 'Name', 'Ticket', 'Cabin'], axis = 1)
x = df_cleaned.iloc[:, :].values # temizlenmiş veriler pclass, sex, age, sibsp, parch, fare, Embarked
y = df.iloc[:, 1].values  # Hayatta kalma durumu survived özniteliği
print(x)
```

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
imputer.fit(x[:, 2].reshape(-1,1))
x[:, 2] = imputer.transform(x[:, 2].reshape(-1,1)).reshape(-1,)
```

```python
imputer = SimpleImputer(missing_values=np.nan, strategy='most_frequent')
imputer.fit(x[:, 6].reshape(-1,1))
x[:, 6] = imputer.transform(x[:, 6].reshape(-1,1)).reshape(-1,)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [6])], remainder='passthrough')
x = np.array(ct.fit_transform(x))

from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
x[:, 4] = le.fit_transform(x[:, 4])
```

```python
# Train Test Verisi Ayırma
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 0)
# random_state = 0 ile veriler sıralı olarak ayrılır. 1 olsaydı rastgele sırada ayırırdı.
```
