```python
import pandas as pd

# Dosya yolunu belirtin
file_path = '/root/.cache/kagglehub/datasets/himanshunakrani/iris-dataset/versions/1/iris.csv'

# CSV dosyasını oku
df = pd.read_csv(file_path)

# İlk 10 satırı görüntüle
print(df.head(10))
```

```python
df.describe()
```

```python
sns.countplot(x=df['species'])
plt.title('Barplot')
```

```python
plt.scatter(df['petal_length'], df['petal_width'], color ='blue')
plt.title('İki Değişkenli Analiz İçin Scatter Plot')
plt.xlabel('petal_length')
plt.ylabel('petal_width')
plt.show()
```

```python
plt.figure(figsize=(10, 10))
sns.scatterplot(x=df['sepal_length'], y=df['sepal_width'], hue=df['species'])
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.scatterplot(x=df['petal_length'], y=df['petal_width'],marker = 's', hue=df['species'], palette = 'copper')
```

```python
plt.hist(df['sepal_width'], bins = 20)
plt.title('Sepal_Width İçin Histogram')
plt.xlabel('sepal_width')
plt.ylabel('Frekans')
plt.show()
```

```python
sns.histplot(x=df["petal_length"], hue=df["species"], multiple="stack")
```

```python
sns.displot(x=df["sepal_width"], bins=25, kde=True)
# bins=25 ifadesi, veri setini 25 aralığa böler ve bu aralıklarda histogram çubukları oluşturur.
# Çekirdek Yoğunluk Tahmini (Kernel Density Estimate - KDE) gösterip göstermeme durumunu belirler.
# Bu, veri setinin olasılık yoğunluğunu yumuşak bir şekilde görselleştirmeye yardımcı olur.
```
